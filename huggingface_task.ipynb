```python
import torch
from transformers import pipeline
from gtts import gTTS
import IPython.display as ipd

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"We are using: {device}")
```

```python
print("Loading Whisper (ASR)...")
asr_pipeline = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-tiny",
    device=device
)

print("Loading Helsinki Translator (En to Ar)...")
translation_pipeline = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-en-ar",
    device=device
)

print("Done! Models are ready.")
```

```python
# Cell 4
text_to_say = "Hello, I am testing this AI model to translate my voice to Arabic."


tts = gTTS(text_to_say, lang='en')
audio_file = "test_audio.mp3"
tts.save(audio_file)

print(f"Created audio file: {audio_file}")
ipd.Audio(audio_file)
```

```python
# Cell 5
def run_full_pipeline(audio_path):
    # 1. Speech to Text
    print(" Listening...")
    asr_out = asr_pipeline(audio_path)
    english_text = asr_out['text']
    print(f"📝 Extracted Text: {english_text}")

    print("-" * 30)

    # 2. Translation
    print(" Translating...")
    trans_out = translation_pipeline(english_text)
    arabic_text = trans_out[0]['translation_text']

    print(f"🇪🇬 Arabic Output: {arabic_text}")
    return arabic_text

# نشغل الدالة على الملف اللي عملناه
result = run_full_pipeline("test_audio.mp3")
```
